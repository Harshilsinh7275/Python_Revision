```python
#if statement
age = int(input("Enter your age: "))
if age > 20:
    print("You are above 20 years old.")
else:
    print("You are 20 years old or younger.")
```

```python
# else 
# The else statement is used in conjunction with an if statement to define a block of code that will be executed if the condition in the if statement evaluates to False. It provides an alternative path of execution when the if condition is not met.
```

```python
# elif
# allows us to check for multiple conditions in sequatial manner.

age = int(input("Enter your age: "))
if age < 13:
    print("You are a child.")       
elif age < 20:
    print("You are a teenager.")
else:
    print("You are an adult.")
```

```python
#nested cinditional statement

age = int(input("Enter your age: "))
if age >= 18:
    print("You are an adult.")
    if age >= 65:
        print("You are a senior citizen.")
    else:
        print("You are not a senior citizen.")
```

```python
num = int(input("Enter a number: "))
if num>=0:
    print("The number is positive.")
    if num%2==0:
        print("The number is even.")
    else:
        print("The number is odd.")
else:
    print("The number is negative.")
```

```python
## Practcal example

# determinie the uf unput year is leap year or usiong the nested conmdition statement
year = int(input("Enter a year: "))
if year %4 ==0:
    if year %100 ==0:
        if year %400 ==0:
            print(year, "is a leap year.")
        else:
            print(year, "is not a leap year.")
    else:
        print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```

```python
# simple calculator using if-else statement
num1 = float(input("Enter a number :"))
num2 = float(input("Enter another number :"))
operation = input("Enter operation (+, -, *, /): ")

if operation == "+":
    result = num1 + num2
elif operation == "-":
    result = num1 - num2
elif operation == "*":
    result = num1 * num2
elif operation == "/":
    if num2 != 0:
        result = num1 / num2
    else:
        result = "Error: Division by zero!"
else:
    result = "Invalid operation!"
print("Result:", result)
```

```python
# determine the grade bazed on the marks using the if else statement
marks = float(input("Enter your marks: "))
if marks >= 90:
    if marks >= 95:
        grade = "A+"
        grade = "Your rank will be in 1 to 5"
elif marks >= 80:
    grade = "B"
elif marks >= 70:
    grade = "C"
elif marks >= 60:
    grade = "D"
elif marks >= 50:
    grade = "E"
else:
    grade = "F"

print("Your grade is:", grade)
```

```python
# loop control statements

## break statement
# the break statement exits the loop prematurely
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
## continues statement
# the continue statement skips the current iteration and moves to the next iteration
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
#nested loops
# a loop inside a loop

for i in range(3):
    for j in range(2):
        print(f"i: {i}, j: {j}")
```

```python
## example - calculate the sum of the foirst n batural numbers using the while and for loop

# while loop

n=10
sum = 0
count = 1

while count<=n:
    sum = sum + count
    count = count + 1

print("Sum of first", n, "natural numbers is:", sum)
```

### list comprehension

Basic syntax                    [expression fro item in iterable]

with conditional logic          [expression for item in iterable if condition]

Nested List comprehensive       [experession for item1 in iterable1 for item2 in iterable2]

```python
## Basic list comprehension syntax

square = [num**2 for num in range(14)]
square
```

```python
## list comprehension withe the condition
list =[]
for i  in range(10):
    if i % 2== 0:
        list.append(i)
print(list)
```

```python
even_numbers =[num for num in range(10) if num % 2 == 0]
print(even_numbers)
```

```python
## Nested list comprehension
lst1 =[1,2,3,4]
lst2 = ['a','b','c','d']

pair = [[i,j] for i in lst1 for j in lst2]

print(pair)
```

```python
## list comrehensions with function call 

words = ["hello", "world", "python", "list", "comprehension"]
lengths = [len(word) for word in words] 
print(lengths)
```
